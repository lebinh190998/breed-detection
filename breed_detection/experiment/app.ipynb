```python
import fastai
import os
```

```python
#/export
from fastai.vision.all import *
import gradio as gr

def label_breed(x): 
    filename = x.split('/')[-1]  # Extract the file name from the path if applicable
    label = ' '.join(word.capitalize() for word in filename.split('_')[:-1])  # Remove the last part (number) after the last underscore
    return label
```

```python
# im = PILImage.create('dog.jpeg')
im = Image.open('golden-retriever.jpeg')
im.thumbnail((192,192))
im
```

```python
#/export
learn = load_learner('breed_identifier_model.pkl')
```

```python
(pred, idx, probs) = learn.predict(im)
print(f"pred: {pred}, index: {idx}, probs: {probs}")
```

```python
path = untar_data(URLs.PETS)/'images'
path
```

```python
def get_file_names(path):
    categories = []
    for root, dirs, files in os.walk(path):
        for file in files:
            label = label_breed(file)
            if label not in categories:
                categories.append(label)
    return categories

categories = get_file_names(path)
categories
```

```python
def classify_image(img):
    (pred, idx, probs) = learn.predict(img)
    print(probs)
    return dict(zip(categories, map(float, probs)))
```

```python
classify_image(im)
```

```python
#/export
image = gr.inputs.Image(shape=(192, 192))
label = gr.outputs.Label()
examples = ['golden-retriever.jpeg']

intf = gr.Interface(fn=classify_image, inputs=image, outputs=label, exmamples=examples)
intf.launch(inline=False)
```
